```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/netflix_titles.csv')
plt.hist(df['release_year'], bins=40, color='skyblue',edgecolor='black')
plt.title('release_year Distribution')
plt.xlabel('release_year')
plt.ylabel('Frequeny')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['date_added'], y=df['release_year'], color='purple')
plt.title('date_added vs. release_year', fontsize=14)
plt.xlabel('date_added', fontsize=12)
plt.ylabel('release_year', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numerical_df = df.select_dtypes(include=['number'])
corr_matrix = numerical_df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```
