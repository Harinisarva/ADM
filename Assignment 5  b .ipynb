```python
import pandas as pd
url='/content/netflix_titles.csv'
data=pd.read_csv(url)
summary=data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['release_year'].dropna(),bins=20,edgecolor='red',color='pink')
plt.title('release_year Distribution')
plt.xlabel('release_year')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(data['duration'].dropna(),bins=20,edgecolor='Black',color='skyblue')
plt.title('duration Distribution')
plt.xlabel('duration')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(data['rating'].dropna(),bins=50,edgecolor='Black',color='skyblue')
plt.title('rating Distribution')
plt.xlabel('rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'Pclass' (Passenger class)
plt.figure(figsize=(6, 4))
data['show_id'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('show_id Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot for 'Age' to visualize outliers and distribution
plt.figure(figsize=(6, 4))
plt.boxplot(data['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of release_year')
plt.ylabel('release_year')
plt.show()
```

```python
import seaborn as sns
# Pairplot to visualize relationships between numerical features
sns.pairplot(data[['release_year', 'show_id', 'rating', 'duration']].dropna())
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data[['release_year', 'release_year', 'release_year', 'release_year']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for 'Survived' (how many survived vs how many did not)
sns.countplot(x='type', data=data, palette='Set2')
plt.title('type Count')
plt.xlabel('type')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plot for 'Age' across 'Pclass'
sns.violinplot(x='release_year', y='release_year', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
